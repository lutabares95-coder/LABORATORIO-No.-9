generame un codigo con las siguientes datos y realiza un análisis simple de las ventas de un pequeño negociocompleto, incluyendo la creación de un DataFrame, el uso de operaciones de NumPy y la visualización de los resultados con Matplotlib.  
Los estudiantes trabajarán con los siguientes datos de ventas mensuales:  Datos de ventas mensuales (en miles)  
data = {“Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"], "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90],}

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Creación del DataFrame
data = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
            "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"],
    "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90]
}

df = pd.DataFrame(data)

# 2. Operaciones con NumPy para el análisis
# Convertimos la columna de ventas a un array de NumPy para cálculos rápidos
ventas_array = df["Ventas"].values

total_ventas = np.sum(ventas_array)
promedio_mensual = np.mean(ventas_array)
mes_max_ventas = df["Mes"][np.argmax(ventas_array)]
crecimiento_total = ((ventas_array[-1] - ventas_array[0]) / ventas_array[0]) * 100

# 3. Visualización de los resultados
plt.figure(figsize=(12, 6))

# Crear gráfico de líneas
plt.plot(df["Mes"], df["Ventas"], marker='o', linestyle='-', color='b', label='Ventas Mensuales')

# Añadir una línea horizontal para el promedio
plt.axhline(y=promedio_mensual, color='r', linestyle='--', label=f'Promedio ({promedio_mensual:.1f}k)')

# Configuración estética
plt.title("Evolución de Ventas Anuales", fontsize=16)
plt.xlabel("Meses", fontsize=12)
plt.ylabel("Ventas (en miles)", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()

# Mostrar el gráfico
plt.show()

# 4. Reporte simple por consola
print("--- Análisis de Ventas ---")
print(f"Ventas totales del año: {total_ventas}k")
print(f"Promedio de ventas mensual: {promedio_mensual:.2f}k")
print(f"Mes con mayor rendimiento: {mes_max_ventas}")
print(f"Crecimiento de Enero a Diciembre: {crecimiento_total:.1f}%")
```

Los estudiantes trabajarán con los siguientes datos de calificaciones:  
 Datos de calificaciones de estudiantes  
 data = {"Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"], "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94], "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93], "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91],}

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de calificaciones de estudiantes proporcionados
data_calificaciones = {
    "Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"],
    "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94],
    "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93],
    "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]
}

# Crear DataFrame de calificaciones
df_calificaciones = pd.DataFrame(data_calificaciones)

print("--- Primeras filas del DataFrame de Calificaciones ---")
display(df_calificaciones.head())
```

### Análisis estadístico de las calificaciones por asignatura usando NumPy y Pandas

```python
# Seleccionar solo las columnas de asignaturas para el análisis numérico
asignaturas = ['Matemáticas', 'Ciencias', 'Historia']

# Calcular la media de las calificaciones por asignatura
media_asignaturas = df_calificaciones[asignaturas].mean()
print("\n--- Media de calificaciones por asignatura ---")
print(media_asignaturas)

# Calcular la calificación mínima por asignatura
min_asignaturas = df_calificaciones[asignaturas].min()
print("\n--- Calificación mínima por asignatura ---")
print(min_asignaturas)

# Calcular la calificación máxima por asignatura
max_asignaturas = df_calificaciones[asignaturas].max()
print("\n--- Calificación máxima por asignatura ---")
print(max_asignaturas)

# Calcular la desviación estándar por asignatura
std_asignaturas = df_calificaciones[asignaturas].std()
print("\n--- Desviación estándar de calificaciones por asignatura ---")
print(std_asignaturas)

# También se puede usar .describe() para un resumen completo
print("\n--- Resumen estadístico completo por asignatura ---")
display(df_calificaciones[asignaturas].describe())
```

### Visualización de los resultados con Matplotlib

```python
# 1. Gráfico de barras de la media de calificaciones por asignatura
plt.figure(figsize=(10, 6))
media_asignaturas.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Calificación Media por Asignatura')
plt.xlabel('Asignatura')
plt.ylabel('Calificación Media')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 2. Gráfico de caja para la distribución de calificaciones por asignatura
plt.figure(figsize=(10, 6))
df_calificaciones[asignaturas].boxplot(grid=False, patch_artist=True, medianprops={'color': 'red'})
plt.title('Distribución de Calificaciones por Asignatura')
plt.xlabel('Asignatura')
plt.ylabel('Calificación')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 3. Calcular la calificación promedio de cada estudiante
df_calificaciones['Promedio General'] = df_calificaciones[asignaturas].mean(axis=1)

# Gráfico de barras para el promedio general de cada estudiante
plt.figure(figsize=(12, 7))
plt.bar(df_calificaciones['Estudiante'], df_calificaciones['Promedio General'], color='salmon')
plt.title('Promedio General de Calificaciones por Estudiante')
plt.xlabel('Estudiante')
plt.ylabel('Promedio General')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("\n--- DataFrame de Calificaciones con Promedio General ---")
display(df_calificaciones)
```
